```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageArr = np.array([10,20,30,30,30,40,50,60,70])
```

```python
weightArr = np.array([20,40,60,57,30,40,50,60,70])
```

```python
plt.plot(ageArr, weightArr, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Weight-Age plot")
```

```python
sampleArr1 = np.linspace(0, 10, 25)
```

```python
sampleArr2 = sampleArr1 ** 3
```

```python
plt.plot(sampleArr1, sampleArr2, "g-*")
```

```python
plt.subplot(1,2,1)
plt.plot(sampleArr1, sampleArr2, "r*-")
plt.subplot(1,2,2)

plt.plot(sampleArr2, sampleArr1, "g--")
```

```python
sampleFigure = plt.figure()
figureAxes = sampleFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(sampleArr1, sampleArr2, "g")
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("Title")
```

```python
sampleFigure = plt.figure()
plot1 = sampleFigure.add_axes([0.1, 0.1, 0.9, 0.9])
plot2 = sampleFigure.add_axes([0.3, 0.5, 0.3, 0.3])

plot1.plot(sampleArr1, sampleArr2, "g")
plot1.set_xlabel("X axis")
plot1.set_ylabel("Y axis")
plot1.set_title("Main plot title")

plot2.plot(sampleArr2, sampleArr1, "b")
plot2.set_xlabel("X axis")
plot2.set_ylabel("Y axis")
plot2.set_title("subplot title")
```

```python
(fig, plots) = plt.subplots(nrows=1, ncols=2)

for plot in plots:
    plot.plot(sampleArr1, sampleArr2, "g")

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6), dpi=100)
plt = fig.add_axes([0.1,0.1,0.9,0.9])

plt.plot(sampleArr1, sampleArr1 ** 2, label="sampleArr ** 2")
plt.plot(sampleArr1, sampleArr1 ** 3, label="sampleArr ** 3")
plt.legend(loc = 0)
```

```python
(fig, plot) = plt.subplots()
plot.plot(sampleArr1, sampleArr2, color="#3A95A8")
plot.plot(sampleArr2,sampleArr1, "#000", alpha=0.7)
```

```python
(fig, plot) = plt.subplots()
plot.plot(sampleArr1, sampleArr1 + 2, color="blue", linewidth=4, linestyle="--")
plot.plot(sampleArr1, sampleArr1 + 3, color="black", linewidth=5, linestyle=":")
plot.plot(sampleArr1, sampleArr1 + 4, color="red", linewidth=3, linestyle="-.")
plot.plot(sampleArr1, sampleArr1 + 5, color="brown", linewidth=3, linestyle="-", marker="o", markerfacecolor="red", markersize = 10)
```
